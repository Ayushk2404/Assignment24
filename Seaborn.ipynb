```python
# Q1
'''
1.Bar Plot:

Uses: Bar plots are used to display and compare the values of different categories or groups. They are often used to show categorical data and their corresponding values.
Common use cases: Showing sales by different product categories, comparing the performance of different teams in a competition, visualizing frequency distributions.

2.Histogram:

Uses: Histograms are used to visualize the distribution of a continuous variable. They display the frequency or count of data points falling within specific intervals (bins).
Common use cases: Analyzing the distribution of exam scores, understanding the distribution of heights in a population, assessing the distribution of income levels.

3.Scatter Plot:

Uses: Scatter plots are used to visualize the relationship between two continuous variables. Each data point on the plot represents a combination of values from the two variables.
Common use cases: Analyzing the correlation between study hours and exam scores, exploring the relationship between temperature and ice cream sales, understanding the connection between advertising spending and sales.

4.Line Plot:

Uses: Line plots (also known as line charts) are used to display the trend or pattern of a variable over a continuous range. They are often used to visualize time series data.
Common use cases: Displaying stock price trends over time, tracking the growth of a company's revenue over several years, showing the temperature variation throughout the day.

5.Heatmap:

Uses: Heatmaps are used to display a matrix of values as a grid of colored cells. They are particularly useful for visualizing correlation matrices and large datasets.
Common use cases: Visualizing correlation between multiple variables, displaying confusion matrices in machine learning, showing the intensity of various factors on a map.'''
```

```python
# Q2
import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.lineplot(x='timepoint', y='signal', data=fmri, hue='event', style='event', markers=True, dashes=False)
```

```python
# Q3 
import seaborn as sns
import matplotlib.pyplot as plt


titanic_data = sns.load_dataset("titanic")


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)  
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")


plt.subplot(1, 2, 2)  
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")


plt.tight_layout()

plt.show()

```

```python
# Q4
import seaborn as sns 
diamonds = sns.load_dataset('diamonds')
sns.histplot(data=diamonds, x = 'price', hue = 'cut', multiple = 'stack')
```

```python
# Q5
import seaborn as sns
import matplotlib.pyplot as plt


iris_data = sns.load_dataset("iris")

sns.set(style="ticks")  
sns.pairplot(iris_data, hue="species")


plt.show()
```

```python
# Q6
import seaborn as sns
import matplotlib.pyplot as plt


flights_data = sns.load_dataset("flights")


flights_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_matrix, cmap='YlGnBu', annot=True, fmt='d')
plt.title("Flights Passengers Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
